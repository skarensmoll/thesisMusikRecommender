```python
def create_spotify_tracks(cr):
    delete = "DROP TABLE IF EXISTS spotify_tracks;"
    cr.execute(delete)
    
    create = "CREATE TABLE spotify_tracks (id INTEGER PRIMARY KEY, id_spotify, name, popularity, album_name, release_date, id_artist, original_artist, original_name)"
    cr.execute(create)
    
def getTrackArtist(artistSongs):

    try:
        cn, cr = create_sqlite_connector()


        ## Add here the for each

        for artistSong in artistSongs:
            artistName = artistSong[0].strip().lower()
            songName = artistSong[1].strip().lower()


            q = 'artist:'+ artistName + ' ' + 'track:'+ songName
            results = spotify.search(q, type='track', limit=3)


            items = results['tracks']['items']

            if items:
                for track in items:
                    album = track['album']
                    artist = track['artists'][0]
                    trackName = track['name']
                    popularity = track['popularity']
                    id_track = track['id']
                    print('inserting..', trackName, artist['name'])
                    
                    sqlite_insert_with_param = """INSERT INTO spotify_tracks
                      (id_spotify, name, popularity, album_name, release_date, id_artist, original_artist, original_name) 
                      VALUES (?, ?, ?, ?, ?, ?, ?, ?);"""

                    data_tuple = (id_track, songName, popularity, album['name'], album['release_date'], artist['id'], artist['name'], trackName)
                    cr.execute(sqlite_insert_with_param, data_tuple)
                    
                    
#                     sql_tab = "INSERT INTO spotify_tracks (id_spotify, name, popularity, album_name, release_date, id_artist, original_artist, original_name) \
#                     VALUES ('%s','%s','%i','%s','%s','%s','%s','%s')" % \
                    
                    
#                     sql = "INSERT INTO spotify_tracks_clone_clone (id_spotify, name, popularity, album_name, release_date, id_artist, original_artist, original_name) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)"
#                     val = (id_track, songName, popularity, album['name'], album['release_date'], artist['id'], artist['name'], trackName)
#                     cr.execute(sql, val)

                    
#                     cr.execute(sql_tab)
                    
    except ValueError:
        print("Error while connecting ", ValueError)
    finally:
        cn.commit()
        cr.close()
        cn.close()
        print("MySQL cn is closed")
```

```python
## Create SpotifyTracks Table
cn, cr = create_sqlite_connector()
create_spotify_tracks(cr)
cn.close()
```
