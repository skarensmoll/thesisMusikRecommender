```python
playlist ='https://open.spotify.com/playlist/2JoWQYjQ2ENstz6lL37GT4?si=ec1a25e014b84e23'
```

```python
def getInfoSongObject(obj):
    x = {}
    x['id'] = obj._raw_track_meta['id']
    x['name'] = obj._raw_track_meta['name']
    x['artist'] = obj._raw_track_meta['artists'][0]['name']
    x['ytlink'] = obj.youtube_link
    x['mainlink'] = 'https://open.spotify.com/track/'+ x['id']
    return x

def getTrackInfo(url):
    x = {}
    info = SongGatherer.from_spotify_url(url)
    x = getInfoSongObject(info)
    return x

def downloadSongYT(song):
    yt = YouTube(song["ytlink"])
    yt.streams.first().download(output_path="../Databases/songs/SpotifyMP3/", filename=song['id']+'.mp3')

def gatherSongsFromPlaylist(url):
    x = []
    info = SongGatherer.from_playlist(url)
    for song in info:
        cleaned = getInfoSongObject(song)
        #downloadSongYT(cleaned)
        x.append(cleaned)
    return x
```

```python
SongsInfo = gatherSongsFromPlaylist(playlist)
```

```python
SongsInfoJSON = json.dumps(SongsInfo)
# SAVE SONGS INFO TO JSON
with open('../Databases/songs/SongsInfo.json', 'w') as outfile:
    outfile.write(SongsInfoJSON)
```
