```python
songs = pd.read_csv('../Experiments/data/batchSongs.csv')
songs = songs.drop(columns=['Unnamed: 0'])
songs['id_track'] = songs['id_track'].apply(lambda x: x.replace('.csv',''))
songs
```

```python
songs = songs[songs.id_track != '5cY8y2XgOfkAh4kSWLFKkz']
songs
```

```python
ids = []
for id in songs['id_track']:
    ids.append(id)
```

```python
features = []
for id in ids:
    feature = sp.audio_features(id)
    features.append(feature)
```

```python
featureDf = pd.DataFrame([s[0] for s in features],columns = ['id_spotify_track', 'id_track', 'danceability', 'energy', 'key',
       'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
       'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature', 'name',
       'popularity', 'album_name', 'release_date', 'year', 'mood'])
featureDf['id_spotify_track'] = ids
```

```python
featureDf
```

```python
for id,value in enumerate(ids):
    song = sp.audio_analysis(value)
    track = song['track']
    featureDf.at[id,'tempo'] = track['tempo']
    featureDf.at[id,'time_signature'] = track['time_signature']
    featureDf.at[id,'key'] = track['key']
    featureDf.at[id,'mode'] = track['mode']
```
